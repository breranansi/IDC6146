## Load Data ##

```python
train_set = tf.keras.utils.image_dataset_from_directory(
    directory='dataset/training_set',
    labels='inferred',
    label_mode='binary',
    batch_size=32,
    image_size=(64, 64)
)
test_set = tf.keras.utils.image_dataset_from_directory(
    directory='dataset/test_set',
    labels='inferred',
    label_mode='binary',
    batch_size=32,
    image_size=(64, 64)
)
```

# Build CNN Model #1 #

(same as homework for now)

```python
tf.random.set_seed(17) #Build it exactly the same every time
```

```python
cnn001=tf.keras.models.Sequential(
    [
    tf.keras.layers.Input(shape=[64,64,3]),
    tf.keras.layers.Rescaling(1.0/255),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=1,activation="sigmoid")
    ]
)
cnn001.summary()
```

# Compile CNN Model #1 #
This is where we change the optimizer.

```python
cnn001.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
%%time
cnnhistory001=cnn001.fit(x = train_set, validation_data = test_set, epochs = 25)
```

```python
plt.plot(cnnhistory001.history['val_loss'])
plt.plot(cnnhistory001.history['loss'])
plt.ylabel('Binary Crossentropy')
plt.xlabel('Epoch')
plt.legend(['test','train'])
plt.grid()
plt.show()
```

```python
plt.plot(cnnhistory001.history['val_accuracy'])
plt.plot(cnnhistory001.history['accuracy'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['test','train'])
plt.grid()
plt.show()
```

# Build CNN Model #4 #
Stochastic gradient with Momentum and Nesterov
Example of non-default optimizer

```python
tf.random.set_seed(17) #Build it exactly the same every time
cnn004=tf.keras.models.Sequential(
    [
    tf.keras.layers.Input(shape=[64,64,3]),
    tf.keras.layers.Rescaling(1.0/255),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=1,activation="sigmoid")
    ]
)
cnn004.summary()
```

```python
%%time
opt004=tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.5,
    nesterov=True,
    weight_decay=None,
    clipnorm=None,
    clipvalue=None,
    global_clipnorm=None,
    use_ema=False,
    ema_momentum=0.99,
    ema_overwrite_frequency=None,
    loss_scale_factor=None,
    gradient_accumulation_steps=None,
    name='vSGD'
)
cnn004.compile(optimizer = opt004, loss = 'binary_crossentropy', metrics = ['accuracy'])
cnnhistory004=cnn004.fit(x = train_set, validation_data = test_set, epochs = 25)
```

```python
plt.plot(cnnhistory001.history['val_loss'])
plt.plot(cnnhistory002.history['val_loss'])
plt.plot(cnnhistory003.history['val_loss'])
plt.plot(cnnhistory004.history['val_loss'])
plt.ylabel('Binary Crossentropy')
plt.xlabel('Epoch')
plt.legend(['ADAM','SGD','RMSprop','vSGD'])
plt.grid()
plt.show()
```

```python
plt.plot(cnnhistory001.history['accuracy'])
plt.plot(cnnhistory002.history['accuracy'])
plt.plot(cnnhistory003.history['accuracy'])
plt.plot(cnnhistory004.history['accuracy'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['ADAM','SGD','RMSprop','vSGD'])
plt.grid()
plt.show()
```
