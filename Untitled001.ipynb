Load Data

```python
train_set = tf.keras.utils.image_dataset_from_directory(
    directory='dataset/training_set',
    labels='inferred',
    label_mode='binary',
    batch_size=32,
    image_size=(64, 64)
)
test_set = tf.keras.utils.image_dataset_from_directory(
    directory='dataset/test_set',
    labels='inferred',
    label_mode='binary',
    batch_size=32,
    image_size=(64, 64)
)
```

Build CNN Model #1

(same as homework for now)

```python
tf.random.set_seed(17) #Build it exactly the same every time
```

```python
cnn001=tf.keras.models.Sequential(
    [
    tf.keras.layers.Input(shape=[64,64,3]),
    tf.keras.layers.Rescaling(1.0/255),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2, strides=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=1,activation="sigmoid")
    ]
)
cnn001.summary()
```

Compile CNN Model

```python
cnn001.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
%%time
cnnhistory001=cnn001.fit(x = train_set, validation_data = test_set, epochs = 25)
```

```python
plt.plot(cnnhistory001.history['val_loss'])
plt.plot(cnnhistory001.history['loss'])
plt.ylabel('Binary Crossentropy')
plt.xlabel('Epoch')
plt.legend(['test','train'])
plt.grid()
plt.show()
```

```python
plt.plot(cnnhistory001.history['val_accuracy'])
plt.plot(cnnhistory001.history['accuracy'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['test','train'])
plt.grid()
plt.show()
```
